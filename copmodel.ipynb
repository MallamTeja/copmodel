```python
!pip install gradio
import gradio as gr
from transformers import AutoModelForSeq2SeqLM, AutoTokenizer

# Load your fine-tuned model
model_name = "anithasoma/nllb-finetuned-telugu"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)
```

```python


# Function for translation
def translate_text(text):
    inputs = tokenizer(text, return_tensors="pt", padding=True, truncation=True)
    outputs = model.generate(**inputs)
    translated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return translated_text

# Gradio UI
with gr.Blocks() as demo:
    gr.Markdown("## 🔠 English to Telugu Translator")
    gr.Markdown("Enter an English sentence, and the model will translate it into colloquial Telugu.")

    with gr.Row():
        with gr.Column():
            input_text = gr.Textbox(label="Enter English Text", placeholder="Type your sentence here...")
        with gr.Column():
            output_text = gr.Textbox(label="Translated Telugu Text", interactive=False)

    translate_button = gr.Button("Translate ✨")

    translate_button.click(translate_text, inputs=input_text, outputs=output_text)

# Launch app
demo.launch()
```
